## IMPORT AND LOAD ##

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df = pd.read_csv('../data/raw/candidates.csv', sep=';') 
print(f"Dataset dimensions: {df.shape}")
display(df.head())
```

## Data types and null values ##

```python
df.info()

nulos = df.isnull().sum()
print("\nNull values per column:\n", nulos[nulos > 0])
```

## Descriptive statistics ##

```python
display(df.describe())

print("\nUnic Values in Code Challenge Score:", sorted(df['Code Challenge Score'].unique()))
print("Unic Values in Technical Interview Score:", sorted(df['Technical Interview Score'].unique()))
```

## Categories Analysis ##

```python
paises_objetivo = [p for p in df['Country'].unique() if p in ['United States', 'USA', 'United States of America', 'Brazil', 'Colombia', 'Ecuador']]
print("Países encontrados para el KPI:", paises_objetivo)

print("\nDistribución de Seniority:")
print(df['Seniority'].value_counts())
```

## Exact duplicates ##

```python
duplicados = df.duplicated().sum()
print(f"Filas duplicadas idénticas encontradas: {duplicados}")
```
